```python
from selenium import webdriver
from urllib.parse import quote_plus
from urllib.request import urlopen
import os

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# (세민) 파이썬 3.10.* / 셀레니움 4.*
```

```python
def save_images(images, save_path):

    for index, image in enumerate(images[:10000]):  # 10000 = 사진 개수
        """ 10000개 안 되면 알아서 끝나더라 """

        src = image.get_attribute('src')
        t = urlopen(src).read()
        file = open(os.path.join(save_path, str(index + 1) + ".jpg"), "wb") # wb가 뭘까.. 화이트밸런스일까....
        file.write(t)
        print("img save " + save_path + str(index + 1) + ".jpg")
```

```python
def create_folder_if_not_exists(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
```

```python
def make_url(search_term):

    # 네이버 이미지 검색
    base_url = 'https://search.naver.com/search.naver?where=image&section=image&query='

    """
    # CCL 상업적 이용 가능 옵션
    end_url = '&res_fr=0&res_to=0&sm=tab_opt&color=&ccl=2' \
              '&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&recent=0&datetype=0&startdate=0&enddate=0&gif=0&optStr=&nso_open=1'
    """

    return base_url + quote_plus(search_term)

# 네이버 형식대로 돼 있음!!!
```

```python
def crawl_images(search_term):

    # url 생성
    url = make_url(search_term)

    # chrome 브라우저 열기(크롬드라이브 오류 생겨서 안 쓰기로 했음~~)
    # 안 돌아가면 """ 돼있는 거 빼고 아래 지우고 크롬드라이브로 해서 해봐
    """
    chromedriver = 'C://chromedriver.exe'
    browser = webdriver.Chrome(chromedriver, options = chrome_options)
    """
    chrome_options = webdriver.ChromeOptions()
    browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options = chrome_options)
    browser.implicitly_wait(3)
    browser.get(url)


    # 브라우저 열기(아래랑 호환이 안 될까봐 넣어놨음..................)
    try:
        WebDriverWait(browser,5).until(
            EC.presence_of_element_located((By.CLASS_NAME, "_image"))
        )

    except Exception as e:
        print(e)


    # 화면 스크롤
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    WebDriverWait(browser,2).until(
    lambda browser: len(browser.find_elements(By.CLASS_NAME, "_image")))
    time.sleep(10)
    """
    이거 때문에 크롬 창 켜지고 스크롤 내려간 상태로 멈출 텐데 저 기다리는 시간 동안
    유저가 직접 화면 리프레시 시켜줘야 이미지 떠서 저장돼!!!! 썸네일 저장하는 거라서
    Q. 왜 썸네일 저장하는 걸로 했는지
    A. 어차피 답답하다고 사이즈 줄일 거 같아서 그냥 썸네일만 잘라오는 걸로...ㅎㅎ
    url 따서 열어가지고 하나하나 저장하는 거 만들면 공유좀
    """


    # 이미지 긁어오기
    images = browser.find_elements(By.CLASS_NAME, "_image")

    # 저장 경로 설정
    # 이미지 저장될 위치: 아무데나 해놔도 검색어로 폴더 만들긴 함 ㄱㅊ
    save_path = "C://Users//seminchoi//Desktop//crawl_images//" + search_term + "//"
    create_folder_if_not_exists(save_path)

    # 이미지 저장
    save_images(images, save_path)

    # 마무리
    print(search_term + " 저장 성공")
    browser.close()
```
